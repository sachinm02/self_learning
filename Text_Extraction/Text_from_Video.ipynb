```python
import cv2
import os
```

```python
# Create a separate folder for our saving frames in our video
if not os.path.exists('image_frames'):
    os.mkdir('image_frames')
```

```python
# Initialising our video
test_video = cv2.VideoCapture("test_video.mp4")
```

```python
# Start our count at 0 for the frames naming
index = 0
while test_video.isOpened():
    ret,frame = test_video.read()
    if not ret:
        break
        
    # creating name for our frames in video
    name = './image_frames/frame_' + str(index) + '.png'
    
    print(f"Extracting frame #{index}")
    cv2.imwrite(name,frame)
    index += 1
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
test_video.release()
cv2.destroyAllWindows()

# NOTE:
# ord('q') returns the Unicode code point of q

# cv2.waitkey(10) returns a 32-bit integer corresponding to the pressed key

# & 0xFF is a bit mask which sets the left 24 bits to zero, because ord() returns a value betwen 0 and 255,
# since your keyboard only has a limited character set

# Therefore, once the mask is applied, it is then possible to check if it is the corresponding key.
```

```python
from PIL import Image
import pytesseract
```

```python
image = Image.open("./image_frames/frame_367.png")
text_data = pytesseract.image_to_string(image,lang='eng')
print("Found this data: \n\n",text_data)
```
